```python
data=[]
for i in df:
    if(i!='Distance in Kilometres'):
        for j in range(len(df[i])):
            data.append([column_cities[j],i,df[i][j]])
```

```python
distances={}
for i in column_cities:
    for j in column_cities:
        distances[(i,j)]=1e8
        distances[(j,i)]=1e8
        
for i in data:
    distances[(i[0],i[1])]=int(i[2])
    distances[(i[1],i[0])]=int(i[2])

for i in column_cities:
    distances[(i,i)]=0
    
```

```python
#heuristic will be: for start node A, next possible node B and goal C, heuristic is dist(A->B) + dist(B->c)
# if B is goal state directly then heuristic is 0
# if no road between any pair then assume distance as 2000 and 3000 resp
```

```python
heuristic=[]
oops=0
for i in column_cities:
    for j in column_cities:
        for k in column_cities:
            if(distances[(i,j)]>(distances[(i,k)]+distances[(k,j)])):
                distances[(i,j)]=distances[(i,k)]+distances[(k,j)]

sett=set()
c=0
for i in distances:
    if(i[0]!=i[1]):
        c+=1
        heuristic.append([i[0],i[1],distances[i]])
```
